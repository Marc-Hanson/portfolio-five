# Data Preperation

Split data in to 'Train-Validation-Test' sets.

```python
import os
import shutil
import random
import joblib


def split_data_to_sets(data_dir, training_ratio, validation_ratio, test_ratio):

    if training_ratio + validation_ratio + test_ratio != 1.0:
        print("Training Ratio, Validation Ratio & Test Ratio must equal 1.")
        return

    labels = os.listdir(data_dir)
    if 'test' in labels:
        pass
    else:
        for folder in ['train', 'validation', 'test']:
            for label in labels:
                os.makedirs(name=data_dir+'/'+folder+'/'+label)

    for label in labels:
        files = os.listdir(data_dir+'/'+label)
        random.shuffle(files)
        train_quantity = int(len(files) * training_ratio)
        validation_quantity = int(len(files) * validation_ratio)
        count = 1
        for file_name in files:
            if count <= train_quantity:
                shutil.move(data_dir+'/'+label+'/'+file_name,
                            data_dir+'/train/'+label+'/'+file_name)
            elif count <= (train_quantity + validation_quantity):
                shutil.move(data_dir+'/'+label+'/'+file_name,
                            data_dir+'/validation/'+label+'/'+file_name)
            else:
                shutil.move(data_dir+'/'+label+'/'+file_name,
                            data_dir+'/test/'+label+'/'+file_name)
            count += 1
        os.rmdir(data_dir + '/' + label)
```

```python
split_data_to_sets(data_dir=f"inputs/leaves_dataset/cherry-leaves",
                   training_ratio=0.7,
                   validation_ratio=0.1,
                   test_ratio=0.2
                   )
```
